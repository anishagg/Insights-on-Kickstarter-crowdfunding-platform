```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
%pylab inline

kick = pd.read_csv("..\data\ks-projects-201801.csv")

print kick.shape
# print kick.info()
print kick.nunique()
```

```python
# variable creation
from datetime import datetime
import math
kick['deadline_date'] = pd.to_datetime(kick['deadline'])
kick['launch_date'] = pd.to_datetime(kick['launched'])
kick['month_launched'] = kick['launch_date'].apply(lambda x: x.month)
kick['year_launched'] = kick['launch_date'].apply(lambda x: x.year)
kick['length_project'] = ((kick['deadline_date'] - kick['launch_date'])/ np.timedelta64(1, 'D')).astype(int)
kick['length_project_M'] = (((kick['deadline_date'] - kick['launch_date'])/ np.timedelta64(1, 'M')).astype(int))+1
kick['log_goal'] = np.where(kick['usd_goal_real'] < 1, 0, np.log10(kick['usd_goal_real']))  
kick['log_pledged'] = np.log10(kick['usd_pledged_real'])
kick['log_backers'] = np.where(kick['backers'] == 0, 0, np.log10(kick['backers']))  
kick['state_new'] = np.where(kick['state'] != 'successful','failure','success')

#drop irrelevant columns
kick1 = kick.drop(['goal','currency','usd pledged'],axis=1)

#drop rows
kick2 = kick1[kick1['state'] != 'live']
kick2 = kick2[kick2['year_launched'] > 2000]

#scale goal amount for categories
cat_goal = kick2.groupby(['category'])['usd_goal_real'].agg(['min','max'])
kick2 = kick2.merge(cat_goal, left_on = "main_category", right_on = cat_goal.index)
kick2['scaled_goal'] = (kick2['usd_goal_real']-kick2['min'])/(kick2['max']-kick2['min'])

#give numeric values to categories to be used later in modeling
category_dict = {
    'Publishing':1,
    'Film & Video':2,
    'Music':3,
    'Food':4,
    'Crafts':5,
    'Games':6,
    'Design':7,
    'Comics':8,
    'Fashion':9,
    'Theater':10,
    'Art':11,
    'Photography':12,
    'Technology':13,
    'Dance':14,
    'Journalism':15
}
kick2['main_category_num'] = kick2['main_category'].apply(lambda x: category_dict[x])

#Colors used in plotting
colorsX = ['#044454']
colorsY = ['#00e269']
colors2 = ['#044454','#00e269']

kick2.shape
kick2.head()
```

```python
# Divide into train and test sets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import classification_report
import sklearn

#get dummy variables
X = pd.get_dummies(kick2[['length_project','year_launched','usd_goal_real','main_category_num','category','country','month_launched']].copy())
y = kick2[['state_new']].copy()
y[y == 'failure'] = 0
y[y == 'success'] = 1

Xarr = X.values
yarr = y['state_new'].tolist()

#Make a train/test split using 30% test size
X_train, X_test, y_train, y_test = train_test_split(Xarr,yarr,test_size=0.3)  
#train random forest
clf = RandomForestClassifier(max_depth=None, min_samples_split=50,random_state=0)
```

```python
#fit the model on train set
clf.fit(X_train,y_train)
#accuracy on train set
scores = cross_val_score(clf, X_train, y_train)
print(scores.mean())

#fit the model on test test
y_predict = clf.predict(X_test)
meanSquaredError=mean_squared_error(y_test, y_predict)
rootMeanSquaredError = math.sqrt(meanSquaredError)
print("RMSE:", rootMeanSquaredError)
print("Accuracy:", sklearn.metrics.accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))
```

```python
importances = clf.feature_importances_
std = np.std([tree.feature_importances_ for tree in clf.estimators_],axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(0,5):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```
